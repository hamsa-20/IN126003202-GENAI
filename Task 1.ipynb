## Q1. What is the difference between "Love" and "love" in NLP?

"Love" and "love" are treated as different tokens in NLP because of case sensitivity. During tokenization, "Love" (capitalized) and "love" (lowercase) are stored separately, which can unnecessarily increase the vocabulary size. Converting all text to lowercase helps maintain consistency and improves model performance.


## Q2. What happens if stopwords are not removed?

If stopwords are not removed, the dataset will contain many common words such as "is", "the", and "and", which usually do not carry significant meaning. This adds noise to the data, reduces model efficiency, increases computational cost, and may negatively impact model accuracy.


## Q3. Provide two real-world scenarios where removing stopwords can be harmful.

1. Sentiment Analysis: Words like "not" are crucial for understanding sentiment. Removing them can completely change the meaning.  
   Example: "not good" → "good" (incorrect interpretation)

2. Question Answering Systems: Words like "who", "what", and "where" are essential for understanding the intent of a question.


## Q4. Difference between stemming and lemmatization

Stemming cuts words to their root form using simple rules. Example: "running" → "run", "studies" → "studi". It may not always produce real words.

Lemmatization converts words to their correct base form using vocabulary and grammar rules. Example: "running" → "run", "better" → "good". It is more accurate but slower than stemming.

```python
#Task 2
import re

def preprocess_text(text):
    if not text or not isinstance(text, str):
        return []

    # Convert to lowercase
    text = text.lower()

    # Remove URLs and emails
    text = re.sub(r'http\S+|www\S+|https\S+|\S+@\S+', '', text)

    # Remove numbers
    text = re.sub(r'\d+', '', text)

    # Handle repeated characters
    text = re.sub(r'(.)\1{2,}', r'\1\1', text)

    # Remove special characters
    text = re.sub(r'[^a-z\s]', '', text)

    # Remove extra spaces
    text = re.sub(r'\s+', ' ', text).strip()

    tokens = text.split()

    # Remove very short tokens except 'no' and 'not'
    filtered_tokens = [
        word for word in tokens
        if len(word) > 2 or word in ["no", "not"]
    ]

    return filtered_tokens


# List of sentences
text = ["I have 2 dogs","This is   good","soooo goooood!!!","WOW!!! This is GREAT!!!","Visit http://example.com now"  ]

# Process each sentence
for sentence in text:
    result = preprocess_text(sentence)
    print("Original:", sentence)
    print("Tokens:", result)
    print("Cleaned:", " ".join(result))
    print("-" * 40)
```

```python
test_sentences = [
    "Get 100% FREE access now!!!",
    "I absolutely looooved this product 😍😍",
    "Worst service ever... 0/10",
    "Call me at 9876543210",
    "This is THE best course!!!",
    "Visit https://openai.com now!",
    "Nooooo this is baaad!!!",
    "OK OK OK I got it",
    "Win $$$ now!!! Limited offer!!!",
    "I am not happy with this"
]

results = []

for sentence in test_sentences:
    tokens, clean = preprocess_text(sentence)
    
    print("Original:", sentence)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print("-"*50)
    
    results.append(tokens)
```

```python
import numpy as np

for i, tokens in enumerate(results):
    total_tokens = len(tokens)
    unique_tokens = len(set(tokens))
    avg_length = np.mean([len(word) for word in tokens]) if tokens else 0

    print(f"Sentence {i+1}")
    print("Total Tokens:", total_tokens)
    print("Unique Tokens:", unique_tokens)
    print("Avg Token Length:", round(avg_length, 2))
    print("-"*40)
```

```python
from collections import Counter

all_tokens = [word for tokens in results for word in tokens]

counter = Counter(all_tokens)

print("Top 10 Frequent Words:")
print(counter.most_common(10))

print("\nTop 5 Least Frequent Words:")
print(counter.most_common()[-5:])
```

```python
def full_pipeline(text_list):
    all_tokens = []
    clean_sentences = []

    for text in text_list:
        tokens, clean = preprocess_text(text)
        all_tokens.extend(tokens)
        clean_sentences.append(clean)

    return {
        "tokens": all_tokens,
        "clean_sentences": clean_sentences
    }
```

```python
edge_cases = [
    "",
    "😍😍😍😍",
    "1234567890"
]

for case in edge_cases:
    tokens, clean = preprocess_text(case)
    print("Input:", case)
    print("Tokens:", tokens)
    print("Cleaned:", clean)
    print("-"*40)
```
